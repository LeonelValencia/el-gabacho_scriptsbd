```python
import pandas as pd

df = pd.read_csv('datasets/anios_marcas.csv')
```

```python
df['DESCRIPCION']
```

```python
import re

# Extract years with a hyphen in the middle from the "DESCRIPCION" column
df['Years'] = df['DESCRIPCION'].apply(lambda x: re.findall(r'\d{2,4}-\d{2,4}', str(x)))

# Flatten the list of years
df['Years'] = df['Years'].apply(lambda x: ', '.join(x) if len(x) > 0 else '')

# Display the dataframe
df
```

```python
df['Years'].unique()
```

```python
years = []

for value in df['Years'].unique():
    groups = value.split(',')
    for group in groups:
        years.append(group.strip())

years = list(set(years))
```

```python
years.remove('170-140')
```

```python
years
```

```python
# Function to convert 2-digit years to 4-digit years
def convert_years(years):
    converted_years = []
    for year in years:
        converted_year = ''
        year2 = year.split('-')
        for y in year2:
            if len(y) == 2:
                start_digit = int(y[0])
                if start_digit >= 5:
                    converted_year += '19' + y + '-'
                else:
                    converted_year += '20' + y + '-'
        converted_years.append(converted_year[:-1])
    return converted_years

# Convert the years in the 'Years' column
years2 = convert_years(years)

# Display the updated dataframe
years2
```
