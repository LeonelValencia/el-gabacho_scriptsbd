```python
import pandas as pd
import re
```

```python
df = pd.read_csv('datasets/importar_cristales.csv')
df = df.iloc[0:26]
df
```

```python
errores_list = []

# Check for missing values in the required columns
required_columns = ['CODIGO', 'NOMBRE', 'EXISTENCIAS', 'PROVEEDOR', 'CATEGORIA', 'MARCA1', 'MODELO1', 'ANIO1']
missing_values = df[required_columns].isnull().any(axis=1)

# Store the rows with errors in the errores variable
filtered_df = df[~missing_values]

# Iterate over the missing values and create error dictionaries
for index, row in df[missing_values].iterrows():
    error_dict = {'fila': index, 'error': ''}
    for column in required_columns:
        if pd.isnull(row[column]):
            error_dict['error'] = f"La columna {column} está vacía"
            break
    errores_list.append(error_dict)

filtered_df
```

```python
# Check and convert the columns COMPRA, MAYOREO, LLEVAR, COLOCADO
for column in ['COMPRA', 'MAYOREO', 'LLEVAR', 'COLOCADO']:
    # Remove the $ symbol and commas, and convert to float
    filtered_df[column] = filtered_df[column].replace({'\$': '', ',': ''}, regex=True)
    
    # Check if the column contains non-numeric values
    non_numeric_values = pd.to_numeric(filtered_df[column], errors='coerce').isna()
    
    # Convert the column to numeric, setting non-numeric values to NaN
    filtered_df[column] = pd.to_numeric(filtered_df[column], errors='coerce')
    
    # Replace NaN values with 0
    filtered_df[column] = filtered_df[column].fillna(0)
    
    # Store the rows with non-numeric values in the errores_list
    error_rows = filtered_df[non_numeric_values].index.tolist()
    for row in error_rows:
        error_dict = {'fila': row, 'error': f"La columna {column} tiene valores no numéricos o esta vacio"}
        errores_list.append(error_dict)

filtered_df.iloc[:,:15]
```

```python
# Convert EXISTENCIAS and CANT_MIN to float and handle NaN values
filtered_df['EXISTENCIAS'] = pd.to_numeric(filtered_df['EXISTENCIAS'], errors='coerce').fillna(0)
filtered_df['CANT_MIN'] = pd.to_numeric(filtered_df['CANT_MIN'], errors='coerce').fillna(0)

# Convert TIPO_MED, PROVEEDOR, CATEGORIA, MARCA1, MODELO1 to uppercase
filtered_df['TIPO_MED'] = filtered_df['TIPO_MED'].str.upper()
filtered_df['PROVEEDOR'] = filtered_df['PROVEEDOR'].str.upper()
filtered_df['CATEGORIA'] = filtered_df['CATEGORIA'].str.upper()
filtered_df['MARCA1'] = filtered_df['MARCA1'].str.upper()
filtered_df['MODELO1'] = filtered_df['MODELO1'].str.upper()

# Check and validate values in TIPO_MED column
invalid_tipo_med = ~filtered_df['TIPO_MED'].isin(['PZA', 'M'])
for row in invalid_tipo_med[invalid_tipo_med].index:
    error_dict = {'fila': row, 'error': f"La columna TIPO_MED tiene un valor inválido"}
    errores_list.append(error_dict)
filtered_df = filtered_df[~invalid_tipo_med]

# Check and validate values in ANIO1 column
invalid_anio1 = ~filtered_df['ANIO1'].str.match(r'^\d{2}-\d{2}$') & ~filtered_df['ANIO1'].str.match(r'^\d{4}-\d{4}$')
for row in invalid_anio1[invalid_anio1].index:
    error_dict = {'fila': row, 'error': f"La columna ANIO1 tiene un valor inválido"}
    errores_list.append(error_dict)
filtered_df = filtered_df[~invalid_anio1]

filtered_df.iloc[:,:15]
```

```python
def validate_anio(anio):
    converted_year = ''
    year2 = anio.split('-')
    for y in year2:
        if len(y) == 2:
            start_digit = int(y[0])
            if start_digit >= 5:
                converted_year += '19' + y + '-'
            else:
                converted_year += '20' + y + '-'
    return converted_year[:-1]
```

```python
productos_list = []

# Iterate over each row of the filtered_df dataframe
for index, row in filtered_df.iterrows():
    vehiculos = []
    
    # Iterate over the different sets of columns
    for i in range(1, 6):
        marca_col = f'MARCA{i}'
        modelo_col = f'MODELO{i}'
        anio_col = f'ANIO{i}'
        
        # Check if the columns are not empty or NaN
        if not pd.isnull(row[marca_col]) and not pd.isnull(row[modelo_col]) and not pd.isnull(row[anio_col]):
            anio = str(row[anio_col]).strip()
            if re.match(r'^\d{2}-\d{2}$', anio):
                converted_year = validate_anio(anio)
                vehiculos.append({'marca': row[marca_col], 'modelo': row[modelo_col], 'anio': converted_year})
            elif re.match(r'^\d{4}-\d{4}$', anio):
                vehiculos.append({'marca': row[marca_col], 'modelo': row[modelo_col], 'anio': anio})
            else:
                errores_list.append({'fila': index, 'error': f'La columna {anio_col} tiene un valor inválido'})
        elif pd.isnull(row[marca_col]) and pd.isnull(row[modelo_col]) and pd.isnull(row[anio_col]):
            None
        else:
            errores_list.append({'fila': index, 'error': f'La columna {marca_col}, {modelo_col} y {anio_col} deben estar llenas o vacías'})
    
    # Add the vehicles to the vehiculos_list if any were found
    if vehiculos:
        productos_list.append({'fila': index, 'vehiculos': vehiculos})

productos_list
```

```python
for index, row in filtered_df.iterrows():
    for producto in productos_list:
        if index == producto['fila']:
            for column in ['CODIGO', 'NOMBRE', 'DESCRIPCION', 'COMPRA', 'MAYOREO', 'LLEVAR', 'COLOCADO', 'EXISTENCIAS', 'CANT_MIN', 'TIPO_MED', 'PROVEEDOR', 'CATEGORIA']:
                producto[column] = row[column]
            producto['vehiculos'] = producto['vehiculos']
```

```python
import time
from mysql.connector import DatabaseError
import mysql.connector

MAX_RETRIES = 3

# Establecer conexión con la base de datos
cnx = mysql.connector.connect(user='root', password='root', host='localhost', port=3307, database='el_gabacho', collation='utf8mb4_general_ci')

# Crear un cursor para ejecutar las consultas SQL
cursor = cnx.cursor()

# Iterar sobre la lista de productos
for producto in productos_list:
    # Consulta para obtener el idCategoria e idProveedor
    categoria_query = f"SELECT idCategoria FROM categorias WHERE nombre = '{producto['CATEGORIA']}'"
    proveedor_query = f"SELECT idProveedor FROM proveedores WHERE empresa = '{producto['PROVEEDOR']}'"

    cursor.execute(categoria_query)
    categoria_result = cursor.fetchone()
    idCategoria = categoria_result[0] if categoria_result else None

    cursor.execute(proveedor_query)
    proveedor_result = cursor.fetchone()
    idProveedor = proveedor_result[0] if proveedor_result else None

    # Verificar si se encontraron los IDs
    if idCategoria is None:
        errores_list.append({'fila': producto['fila'], 'error': f'No se encontró idCategoria con la categoría {producto["CATEGORIA"]}'})
        continue
    if idProveedor is None:
        errores_list.append({'fila': producto['fila'], 'error': f'No se encontró idProveedor con el proveedor {producto["PROVEEDOR"]}'})
        continue

    # Determinar el idUnidadMedida
    idUnidadMedida = 1 if producto['TIPO_MED'] == 'PZA' else 2 if producto['TIPO_MED'] == 'M' else None

    if idUnidadMedida is None:
        errores_list.append({'fila': producto['fila'], 'error': 'Valor inválido para TIPO_MED'})
        continue

    # Parámetros para el procedimiento almacenado
    params = [
        idCategoria,
        idUnidadMedida,
        producto['CODIGO'],
        producto['NOMBRE'],
        producto['DESCRIPCION'],
        producto['EXISTENCIAS'],
        producto['CANT_MIN'],
        producto['COMPRA'],
        producto['MAYOREO'],
        producto['LLEVAR'],
        producto['COLOCADO'],
        idProveedor,
        3,  # idUsuario
        0  # Parámetro de salida p_idInventario
    ]

    for i in range(MAX_RETRIES):
        try:
            print(f"intento {i} del producto {producto['fila']}")
            # Llamar al procedimiento almacenado
            result_args = cursor.callproc('proc_insertar_producto', params)
            break
        except DatabaseError as e:
            if 'Lock wait timeout exceeded' in str(e):
                time.sleep(5)  # Espera antes de reintentar
            else:
                raise

    # El valor de salida se encuentra en result_args[-1]
    idInventario = result_args[-1]
    print(f"ID de Inventario generado: {idInventario}")

    if idInventario == 0:
        errores_list.append({'fila': producto['fila'], 'error': 'No se generó el idInventario'})
        continue
    
    idModeloAnios_list = []
    for vehiculo in producto["vehiculos"]:
        # Define the SQL query to get the idMarca
        marca_query = f"SELECT idMarca FROM marcas WHERE nombre = '{vehiculo['marca']}'"
        
        # Execute the query to get the idMarca
        cursor.execute(marca_query)
        idMarca = cursor.fetchone()[0]
        
        # Check if idMarca exists
        if idMarca is None:
            # Add the error to errores_list and move to the next vehicle
            errores_list.append({'fila': producto['fila'], 'error': f'No se encontró idMarca con la marca {vehiculo["marca"]}'})
            continue

        # Retrieve the idModelo for the given model name "VITARA" and idMarca
        cursor.execute('SELECT idModelo FROM modelos mo JOIN marcas ma ON mo.idMarca = ma.idMarca WHERE mo.nombre = %s AND mo.idMarca = %s', (vehiculo['modelo'],idMarca,))
        modelo_result = cursor.fetchone()
        idModelo = modelo_result[0] if modelo_result is not None else None

        # Check if idModelo exists
        if idModelo is None:
            errores_list.append({'fila': producto['fila'], 'error': f'No se encontró idModelo con el modelo {vehiculo["modelo"]}'})
            continue

        # Retrieve the idAnio for the given start year "1989" and end year "1998"
        anios = vehiculo['anio'].split('-')
        cursor.execute(f'SELECT idAnio FROM anios WHERE anioInicio = "{anios[0]}" AND anioFin = "{anios[1]}"')
        anio_result = cursor.fetchone()
        idAnio = anio_result[0] if anio_result is not None else None

        # Check if idAnio exists
        if idAnio is None:
            errores_list.append({'fila': producto['fila'], 'error': 'No se encontró idAnio'})
            continue

        # Retrieve the idModeloAnio for the given idModelo and idAnio
        cursor.execute('SELECT ma.idModeloAnio FROM modeloAnios ma JOIN modelos m ON ma.idModelo = m.idModelo WHERE ma.idAnio = %s AND ma.idModelo = %s', (idAnio, idModelo))
        modelo_anio_result = cursor.fetchone()
        idModeloAnio = modelo_anio_result[0] if modelo_anio_result is not None else None

        # Check if idModeloAnio exists
        if idModeloAnio is None:
            # Create the idModeloAnio using the stored procedure proc_modelo_anio
            cursor.callproc('proc_insertar_modelos', (idModelo, anios[0], anios[1], False))
            cursor.execute('SELECT ma.idModeloAnio FROM modeloAnios ma JOIN modelos m ON ma.idModelo = m.idModelo WHERE ma.idAnio = %s AND ma.idModelo = %s', (idAnio, idModelo))
            idModeloAnio = cursor.fetchone()[0]
            
        idModeloAnios_list.append(str(idModeloAnio))

    # Execute the stored procedure proc_modeloanios_con_autoparte
    idModeloAnios_string = ','.join(idModeloAnios_list)
    print(f"ID de ModeloAnios generado: {idModeloAnios_string}")
    cursor.callproc('proc_relate_producto_modeloanios', (idInventario, idModeloAnios_string))

# Commit the changes and close the cursor and connection
cnx.commit()
cursor.close()
cnx.close()
```
